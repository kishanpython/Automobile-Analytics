```python
# import neccessary libs
import os
import psycopg2
import pandas as pd
from sqlalchemy import create_engine
```

```python
# reading the data to store it in database
df = pd.read_csv("truck1.csv")
```

```python
# taking 50 rows and 10 columns for sample storing of data in database
new_df =  df.iloc[:50,:10]
```

```python
# cheking of shape
new_df.shape
```

```python
# converting the date field from string to datetime object
new_df["Date"] = pd.to_datetime(new_df["Date"], format="%d-%m-%Y")
# change to date only
new_df['Date'] = new_df['Date'].dt.date
```

```python
# listing of all datatypes of columns for making schema
new_df.dtypes
```

```python
# data storage function
def data_store_psql(complete_details,vehicle_name):
    # connection string 
    conn = psycopg2.connect(database="vehicledb", user = "postgres", password = "root",)
    # cursor 
    cur = conn.cursor()
    # schema of our table
    comp_schema = """
        date DATE NOT NULL,
        latitude FLOAT,
        longitude FLOAT,
        altitude FLOAT,
        rpm FLOAT,
        driver_demand_torque_per INTEGER,
        engine_load_per INTEGER,
        engine_torque_mode INTEGER,
        tps_per FLOAT,
        percent_load_curret_speed INTEGER
        """
    # table name
    table_name = vehicle_name
    try:
        # if table not exist create new one else add data into it
        command = "CREATE TABLE IF NOT EXISTS %s (%s) "%(table_name, comp_schema)
        # execute the above command to create and check table in our database
        cur.execute(command)
        # looping through data and inserting into the table
        for info in complete_details:
            # query to insert the data
            insert_to = "INSERT INTO %s (date,latitude,longitude,altitude,rpm,driver_demand_torque_per,engine_load_per,engine_torque_mode,tps_per,\
            percent_load_curret_speed) VALUES (%%s, %%s, %%s, %%s, %%s, %%s, %%s, %%s, %%s, %%s)" % table_name
            # data to store into table
            data = info
            # execute above command
            cur.execute(insert_to,(data))
        # commmiting the changes    
        conn.commit()
        # closeing the connection
        conn.close()
        return 0
    except:
        print("NO TABLE FOUND!")
        return 1
```

```python
# data in db
data = new_df.values.tolist()
# call the func
data_store_psql(data, 'vehicle_test')
```
